# 가상화폐 가격 와 인터넷 글 갯수의 관계

- 가상화폐의 가격과 인터넷글의 갯수, 구글 트렌드를 통하여 
- 이를 통해 인터넷 사용자들이 비트코인의 언급량을 통해 투자 시기를 가늠해 보도록 하자.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
file_path = r'market-price.json'
bitcoin_df = pd.read_json(file_path)

print(bitcoin_df)
```

```python
# market-price 열의 값만 추출하여 리스트로 저장
market_prices = bitcoin_df['market-price'].tolist()

print(market_prices[:5])
```

```python
# 'x'와 'y' 값을 분리하여 리스트로 저장
x_values = [d['x'] for d in market_prices]
y_values = [d['y'] for d in market_prices]

print(x_values[:3])
print(y_values[:3])
```

```python
# 'x' 값을 datetime 객체로 변환 (마이크로 초로 표현되어 있어 년 월로 표현하기)
x_datetimes = [pd.to_datetime(x, unit='ms') for x in x_values]
x_datetimes[:3]
```

```python
# 그래프 그리기
fig, ax = plt.subplots(figsize=(12, 6)) 
ax.plot(x_datetimes, y_values)
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.title('Bitcoin Price Over Time')
plt.show()
```

```python
# 그래프 그리기
fig, ax = plt.subplots(figsize=(12, 6)) # 그래프의 크기를 12인치 x 6인치로 설정하고, fig와 ax 객체를 생성합니다.
ax.plot(x_datetimes, y_values) # ax 객체에 x_datetimes(x축 값)와 y_values(y축 값)를 이용하여 선 그래프를 그립니다.
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d')) # x축의 날짜 형식을 '%Y-%m-%d' 형식으로 설정합니다.
plt.xlabel('Date') # x축의 레이블을 'Date'로 설정합니다.
plt.ylabel('Price (USD)') # y축의 레이블을 'Price (USD)'로 설정합니다.
plt.title('Bitcoin Price Over Time') # 그래프의 제목을 'Bitcoin Price Over Time'으로 설정합니다.
plt.grid(True)
plt.show() # 그래프를 화면에 출력합니다.
```

```python
import ast

# 문자열로 받아 와서 ast를 사용함
with open('comunity_howmany_bitcoin.txt', 'r') as file:
    data = file.read()

data_list = ast.literal_eval(data)
print(data_list)
```

```python
from datetime import datetime as dt

datetime_object = []
datetime_values = []

# 데이터 순회하면서 year, month, value 추출하여 datetime 객체 생성하고 리스트에 추가
for date_string, value in data_list.items():
    year, month = map(int, date_string.split('-'))
    datetime_object.append([year, month, 1])  # 일자는 1일로 설정
    datetime_values.append(value)

# x축 전처리
dates = [datetime(year, month, day) for year, month, day in datetime_object]

print(dates)
print(datetime_values)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(dates, datetime_values)
plt.xlabel('Date')
plt.grid(True)
plt.show()
```
