```python
def group_same_values(lst):
    grouped_list = []
    current_group = []

    # 리스트 순회
    for item in lst:
        if not current_group or item == current_group[0]:
            # 현재 그룹이 비어 있거나, 현재 아이템이 현재 그룹의 첫 번째 아이템과 같으면
            current_group.append(item)  # 현재 그룹에 아이템 추가
        else:
            # 그렇지 않으면 현재 그룹을 그룹된 리스트에 추가하고 새 그룹 시작
            grouped_list.append(current_group)
            current_group = [item]  # 새로운 그룹 시작

    # 마지막 그룹 추가
    if current_group:
        grouped_list.append(current_group)

    return grouped_list

# 예시 리스트
my_list = [1, 1, 2, 3, 3, 3, 4, 5, 5]

# 같은 값을 가진 요소들을 묶은 리스트 생성
result = group_same_values(my_list)
print(result)

```
