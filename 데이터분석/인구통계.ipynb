```python
import csv
import matplotlib.pylab as plt

f = open('age.csv', 'r',encoding='cp949')
data = csv.reader(f, delimiter = ',')

header = next(data)

result = []

what_you_want = input('찾고 싶은 지역의 이름을 알려주세요.')

for row in data:
    if what_you_want in row[0]:
        for i in row[3:]:
            i = int(i.replace(',',''))
            result.append(int(i))

print(result)

f.close()


plt.style.use('ggplot')
plt.plot(result)
plt.show()
```

```python
import csv
import matplotlib.pylab as plt

f = open('gender.csv', 'r',encoding='cp949')
data = csv.reader(f, delimiter = ',')

header = next(data)

male = []
female = []

#what_you_want = input('찾고 싶은 지역의 이름을 알려주세요.')

for row in data:
    if '소공동' in row[0]:
        for i in range(0,101):
            male.append(-int(row[i+3]))
            #print(int(row[i+3]))
            female.append(int(row[-(i+1)]))
            #print(int(row[-(i+1)]))

female.reverse()
#print(female)

plt.title('population distribution by gender')
plt.style.use('ggplot')
plt.barh(range(101),male, label = 'MAN')
plt.barh(range(101),female, label = 'WOMAN')
plt.legend()
plt.show()


f.close()
```

```python
import csv
import matplotlib.pylab as plt

f = open('gender.csv', 'r',encoding='cp949')
data = csv.reader(f, delimiter = ',')

header = next(data)

male = 0
female = 0

total = []

dongname = input('찾고 싶은 지역의 이름을 알려주세요')

for row in data:
    if dongname in row[0]:
        for i in range(101):
            male += int(row[i+3])
            female += int(row[i + 106])
        break;


#plt.pie([10,20])
#plt.pie([10,20,30,40])
total.append(male)
total.append(female)
print(total)

color = ['crimson', 'darkcyan']
plt.title('dongname')

plt.pie(total, labels=['male','female'], colors=color, autopct="%.1f%%",startangle=90)
plt.legend()
plt.show()
```
